```python
def purchase(df):
    if (df['ONEOFF_PURCHASES'] == 0 ) & (df['INSTALLMENTS_PURCHASES']==0):
        return 'none'
    if (df['ONEOFF_PURCHASES']> 0 ) & (df['INSTALLMENTS_PURCHASES']>0):
        return 'both_oneoff_installment'
    if (df['ONEOFF_PURCHASES'] >0 ) & (df['INSTALLMENTS_PURCHASES']==0):
        return 'oneoff'
    if (df['ONEOFF_PURCHASES'] == 0 ) & (df['INSTALLMENTS_PURCHASES']>0):
        return 'installment'
```

```python
#create new categorical column using function defined above
df['purchase_type']=df.apply(purchase, axis=1)
df.head()
```

##### Apply principle component analysis(PCA)

```python
from sklearn.decomposition import PCA
pc = PCA(n_components=6)
df_pca=pc.fit(df_scaled)
df_pca
```

```python
sum(df_pca.explained_variance_ratio_)
```

```python
var_ratio= {}
for i in range(2,23):
    pc = PCA(n_components=i)
    df_pca = pc.fit(df_scaled)
    var_ratio[i] = sum(df_pca.explained_variance_ratio_)
```

##### Selecting clusters based on Silhouette score

The Silhouette score in the K-Means clustering algorithm is between -1 and 1. This score represents how well the data point has been clustered, and scores above 0 are seen as good, while negative points mean your K-means algorithm has put that data point in the wrong cluster.

```python
from sklearn.metrics import silhouette_samples, silhouette_score
k_range = range(2,21)
for k in k_range:
    km = KMeans(n_clusters=k)
    km.fit(reduced_df)
    score=(silhouette_score(reduced_df,km.labels_))
    print("Cluster  no.{} :".format(k),  score)
```
